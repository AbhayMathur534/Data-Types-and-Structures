2. Difference between mutable and immutable datatypes

Mutable Data Types:-> Those data structures whose values can be changed during the run-time or there after are Mutable Data Types or in simpler words the dynamic data types are termed to be as Mutable Data Types. Their values can be changed by indexing the specific elements and replacing them with a corresponding new value.

Immutable Data Types:-> Those data structures whose values can not be changed neither during run-time nor there after are Immutable Data Types. They are specifically can be described as the static data types. Since they are static data types, their values can not be changed by any way possible.

An example of mutable data types is LIST, DICTIONARY data structure because in both the data structure their values can be accessed and changed correspondingly.

An example of immutable data types is TUPLE because in TUPLE we can't change the corresponding value once stored.

4. Dictionary data structure in python

Dictionary is also a heterogenous data structure.

In a dictionary, a key must always be a unique value that should not be any other data structure apart from string or integer but the value can be any value such as string, integer, list, tuple or any other data type.

A dictionary is represented by **curly brackets {}** such that value is stored like this:->

Dict = {"Name":"Abhay", "Course":"DS", "Age":21}

Here, the keys are **Name, Course and Age** and the values are **Abhay, DS and 21**.

So, this was all about dictionary in python.

20. Difference between list and dictionary in terms of data retrieval.

When we talk about data retrieval we know that in list we retrieve data by using the index number alloted to that particular element.

For example:->

lst = [1, 2, 3, 4, 5]
lst[0]

It will result into 1

So, we are retrieving the data using the index number alloted to it.

But when we talk about in dictionary, then retrieval is done using the keys of that corresponding dictionary to get the corresponding values of that keys.

For example:->

dict = {"apple": "1 Kg", "banana":"2 Dozen", "mango":"1 Kg"}
dict["apple"]

It will result into '1 Kg'.

So, we are retrieving the data(value in this scenario) using the corresponding key of the dictionary.

So, this was all about it.
